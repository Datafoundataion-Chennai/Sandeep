```python
import numpy as np
import matplotlib.pyplot as plt


days = np.arange(1, 11)  
stock_price = [150, 152, 149, 155, 160, 162, 158, 165, 170, 175]  # In USD
trading_volume = [1.2, 1.4, 1.1, 1.8, 2.0, 2.5, 2.2, 2.7, 3.0, 3.5]  # In million shares

fig, ax1 = plt.subplots()

ax1.set_xlabel("Days")
ax1.set_ylabel("Stock Price (USD)", color='tab:blue')
ax1.plot(days, stock_price, 'o-', color='tab:blue', label="Stock Price")
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel("Trading Volume (Million Shares)", color='tab:red')
ax2.plot(days, trading_volume, 's--', color='tab:red', label="Trading Volume")
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title("Stock Price & Trading Volume Over 10 Days")
ax1.grid(True, linestyle=':', alpha=0.5)

plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
rent = [1200, 1200, 1200, 1200, 1250, 1250]
food = [400, 420, 450, 480, 500, 520]
transport = [100, 120, 130, 110, 140, 160]
entertainment = [150, 200, 220, 180, 250, 300]
rent=np.array(rent)
rent=np.array(food)
rent=np.array(transport)
rent=np.array(entertainment)

plt.bar(months, rent, color='Red')
plt.bar(months, food, bottom=rent, color='blue')
plt.bar(months, transport, bottom=rent+food, color='green')
plt.bar(months, entertainment, bottom=rent+food+transport, color='orange')

plt.xlabel('months')
plt.ylabel('Monthly expenses')
plt.legend((['rent','food','entertainment','transport']))
plt.show()

```

```python
from collections import OrderedDict

def Canteen_Supply_Tracker(requests):
    # item_counter = Counter()
    ordered_items = OrderedDict()
    
    for item,quantity in requests:
        if item in ordered_items:
             
            ordered_items[item] += quantity
        else :
            ordered_items[item] = quantity
    return ordered_items
            

request= [
    ("bread", 5),
    ("milk", 2),
    ("bread", 3),
    ("cigarettes", 4),
    ("milk", 1)
]
result = Canteen_Supply_Tracker(request)
print(result)
```
